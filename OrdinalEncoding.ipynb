```python
customer = customer.iloc[:,2:]
```

```python
customer.head()
```

```python
#Primarily , the important step is to do train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(customer.iloc[:,:2],customer.iloc[:,-1],test_size=0.2,random_state=42)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
```

```python
oe.fit(X_train)
```

```python
X_train = oe.transform(X_train)
X_test = oe.transform(X_test)
```

```python
X_train
```
