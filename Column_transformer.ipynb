```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/drive/MyDrive/covid_toy.csv")
```

```python
df.isnull().sum()
#fever has missing data
```

```python
from sklearn.impute import SimpleImputer
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['has_covid']),df['has_covid'],test_size=0.2,random_state=42)
```

```python
X_train.head()
```

```python
#using Simple Imputer to fill the missing values by means of feature:fever
#replacing missing data using mean of feature:fever
#by default Simple Imputer replaces missing data by means of that feature
si = SimpleImputer()
X_train_fever = si.fit_transform(X_train[['fever']])
X_test_fever = si.fit_transform(X_test[['fever']])
```

```python
X_train_fever
```

```python
df.cough.value_counts()
```

```python
#Ordinal Encoding
oe = OrdinalEncoder(categories=[['Mild','Strong']])
```

```python
X_train_cough = oe.fit_transform(X_train[["cough"]])
X_train_cough.shape
```

```python
X_test_cough = oe.fit_transform(X_test[["cough"]])
```

```python
#OneHotEncoding
ohe = OneHotEncoder(drop="first",sparse_output=False)
```

```python
X_train_gender_city = ohe.fit_transform(X_train[['gender','city']])
```

```python
X_test_gender_city = ohe.fit_transform(X_test[['gender','city']])
X_train_gender_city.shape
```

```python
df.city.value_counts()
```

```python
df.head()
```

```python
#Extracting age
X_train_age = X_train.drop(columns=['gender','fever','cough','city']).values
X_test_age = X_test.drop(columns=['gender','fever','cough','city']).values
```

```python
X_train_age
```

```python
#Merging all the transformed columns
X_train_transformed = np.concatenate((X_train_age,X_train_fever,X_train_gender_city,X_train_cough),axis=1)
X_test_transformed = np.concatenate((X_test_age,X_test_fever,X_test_gender_city,X_test_cough),axis=1)
```

```python
X_train_transformed
```

Using Column Transformer

```python
from sklearn.compose import ColumnTransformer
transformer = ColumnTransformer(transformers=[
    ("tnf1",SimpleImputer(),['fever']),
    ("tf2",OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
    ("tnf3",OneHotEncoder(sparse_output=False,drop='first'),['gender','city'])
],remainder="passthrough")
```

```python
transformer.fit(X_train)
```

```python
transformer.transform(X_train).shape
```

```python
transformer.transform(X_test).shape
```
